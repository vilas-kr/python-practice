```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 1 Bar chart of the popular of programming language
programming_language = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(programming_language, popularity)
plt.title("Popular Programming Language")
plt.xlabel("Language")
plt.ylabel("Popularity")
plt.show()
```

```python
# 2 Horizontal bar of popular programming language
plt.title("Popular Programming Language")
plt.barh(programming_language, popularity)
plt.xlabel("Popularity")
plt.ylabel("Language")
plt.show()
```

```python
# 3 Popularity of programming language using unifrom color
plt.bar(programming_language, popularity, color="green")
plt.title("Popular Programming Language")
plt.xlabel("Language")
plt.ylabel("Popularity")
plt.show()
```

```python
# 4 Different color for each bar
plt.bar(programming_language, popularity, color=['green', 'red', 'yellow', 'orange', 'pink', 'gray'])
plt.title("Popular Programming Language")
plt.xlabel("Language")
plt.ylabel("Popularity")
plt.show()
```

```python
# 5 Attach a text label above the bar chart
bars = plt.bar(programming_language, popularity, color="green")
plt.title("Popular Programming Language")
plt.xlabel("Language")
plt.ylabel("Popularity")
plt.bar_label(bars, color="red", padding=2)
plt.show()
```

```python
# 6 Add border to barchard
plt.bar(programming_language, popularity, color="green", edgecolor="blue", linewidth=2)
plt.title("Popular Programming Language")
plt.xlabel("Language")
plt.ylabel("Popularity")
plt.show()
```

```python
# 7 Specifying the position of each bar plot
pos = np.arange(len(programming_language))
plt.bar(pos, popularity)
plt.xticks(pos, programming_language)
plt.title("Popular Programming Language")
plt.xlabel("Language")
plt.ylabel("Popularity (%)")
plt.show()
```

```python
# 8 Select the width of each bar and their positions 
pos = np.arange(len(programming_language)) * 1.1
plt.bar(programming_language, popularity, width=[1, 0.5, 0.7, 1.5, 0.3, 1])
plt.xticks(pos, programming_language)
plt.title("Popular Programming Language")
plt.xlabel("Language")
plt.ylabel("Popularity (%)")
plt.show()
```

```python
# 9 Increase bottom margin
plt.figure(figsize=(5, 5))
plt.bar(programming_language, popularity)
plt.title("Popular Programming Language")
plt.xlabel("Language")
plt.ylabel("Popularity (%)")
plt.subplots_adjust(bottom=0.4)
plt.show()
```

```python
# 10 Multiple values on the same chart
groups = ['G1', 'G2', 'G3', 'G4', 'G5']
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

pos = np.arange(len(groups))
bar_width = 0.35

plt.bar(pos - bar_width/2, men,
        width=bar_width,
        label='Men',
        edgecolor='black')

plt.bar(pos + bar_width/2, women,
        width=bar_width,
        label='Women',
        edgecolor='black')

plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by Group and Gender')

plt.xticks(pos, groups)
plt.legend()
plt.show()
```

```python
# 11 Create a bar plot using DataFrame
data = {
    'a': [2,4,6,8,10],
    'b': [4,2,4,2,2],
    'c': [8,3,7,6,4],
    'd': [5,4,4,4,3],
    'e': [7,2,7,8,3]
}
df = pd.DataFrame(data)
df.plot(kind='bar')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Bar Plot from DataFrame')
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(0.86,1))
plt.show()
```

```python
# 12 Bar chart with error bars

mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]

x_pos = np.arange(len(mean_velocity))
plt.figure(figsize=(8, 5))

plt.bar(x_pos,
        mean_velocity,
        yerr=std_deviation,
        capsize=6,
        edgecolor='black')

plt.xlabel('Sample')
plt.ylabel('Mean Velocity')
plt.title('Mean Velocity with Error Bars')
plt.xticks(x_pos, ['A', 'B', 'C', 'D'])
plt.show()
```

```python
# 13 Error bar with text

mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]

x_pos = np.arange(len(mean_velocity))

bars = plt.bar(x_pos,
               mean_velocity,
               yerr=std_deviation,
               capsize=6,
               edgecolor='black')

plt.xlabel('Sample')
plt.ylabel('Mean Velocity')
plt.title('Mean Velocity with Error Bars')

plt.xticks(x_pos, ['A', 'B', 'C', 'D'])

# Attach integer value labels above each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2,
             height,
             f'{int(height)}',
             ha='right',
             va='bottom')
plt.show()
```
