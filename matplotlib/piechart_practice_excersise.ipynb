```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 1 Pie chart of popular programming language
programming_language = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.pie(popularity, autopct="%1.2f%%")
plt.legend(labels=programming_language,bbox_to_anchor=(0.9,0.8))
plt.show()
```

```python
# 2 Pie chart with title
programming_language = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.pie(popularity, autopct="%1.2f%%")
plt.legend(labels=programming_language,bbox_to_anchor=(0.9,0.8))
plt.title("Popular Programming Language", loc='center')
plt.show()
```

```python
# 3 Read Data from csv
df = pd.read_csv('medal.csv')
plt.pie(df['gold_medal'], autopct="%1.2f%%")
plt.legend(labels=df['country'], bbox_to_anchor=(0.9,0.9))
plt.title("Five Most Successful Countries in 2016 Summer Olympics")
plt.show()
```
