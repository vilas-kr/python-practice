```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sales.csv')
```

```python
plt.plot(data['Sales'])
plt.ylabel("sales")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data['Sales'], data['Day'])
plt.xlabel("Days")
plt.ylabel("Sales")
plt.minorticks_on()
plt.xticks([10000, 20000, 30000, 40000])
plt.yticks([x for x in range(0, 21, 2)])
plt.title("Sales data analysis")
plt.grid(which="both", linewidth='.08')
plt.show()
```

```python
# Visualize sales and profit
plt.plot(data['Day'], data['Sales'], color='red', marker = "*",  mec="blue", mfc="yellow", markersize=10, label="Sales")
plt.plot(data['Day'], data['Profit'], color='blue', marker=">", mfc="pink", linestyle="-", linewidth=0.5, label="Profit")
plt.xlabel("days", fontdict={'family':'serif','color':'blue','size':20}, loc="center")
plt.ylabel("sales", loc="bottom")
plt.legend(loc="upper right", bbox_to_anchor=(1.0, 1.2))
plt.title("Profit and Sales")
plt.grid(color='black', axis="x", linestyle = ':', linewidth = 0.5)
plt.show()
```

```python
plt.bar(data['Month'], data['Sales'], color="brown", width=0.5, label="sales", edgecolor="blue", alpha=0.5, hatch="/", align='edge')
plt.bar(data['Month'], data['Profit'], color="yellow", width=0.5, label="profit", edgecolor="blue", hatch="/", align='center')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['Sales'], data['Profit'], 
            s=data['Orders'],          # marker size proportional to Orders
            c=data['Discount'],        # color by Discount
            cmap='coolwarm',         # colormap
            alpha=0.8,
            edgecolors='black',
            linewidths=1.5,
            label='Daily Sales')
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs Profit Scatter")
plt.colorbar(label='Discount %')
plt.legend()
plt.show()
```

```python
# Sales and Profit on regions
import matplotlib.patches as mpatches
plt.scatter(data['Profit'], data['Sales'], s=data['Orders'], color=data['Region'].map({"North":"yellow", "South":"red", "East":"blue", "West":"orange"}))
colors = {"North":"yellow", "South":"red", "East":"blue", "West":"orange"}
handles = [mpatches.Patch(color=color, label=region) for region, color in colors.items()]
plt.legend(handles=handles, title="Region")
plt.xlabel("Profit")
plt.ylabel("Sales")
plt.title("Profit vs Sales vs Region")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8,5))

plt.hist(data['Sales'], 
         bins=5, 
         color='steelblue', 
         edgecolor='black', 
         linewidth=1.5,
         alpha=0.8,
         rwidth=0.9,
         label='Sales Distribution')

plt.xlabel("Sales")
plt.ylabel("Count")
plt.title("Histogram of Daily Sales")
plt.legend()
plt.grid(True)
plt.show()

```
