```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1 & 2 Draw line with suitable x and y axis
sales = {'days' : ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
        'sales': [120, 150, 100, 170, 200]}
df = pd.DataFrame(sales)
plt.plot(df['sales'], df['days'])
plt.title("Sales in a Week")
plt.xlabel("Sales (Rupees)")
plt.ylabel("Days")
plt.show()
```

```python
# 3 Read data from txt file and plot line chart
df = pd.read_csv("test.txt", sep=" ")
plt.plot(df['1'], df['2'])
plt.title("Data readd from test file")
plt.xlabel("X axis")
plt.ylabel(" Y axis")
plt.show()
```

```python
# 4 Line chart on Financial data
df = pd.read_csv('fdata.csv')
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%y")
plt.title("Financial data from 3rd Oct to 7th Oct")
plt.plot(df['Open'], df['Date'], label="Open")
plt.plot(df['High'], df['Date'], label="High")
plt.plot(df['Low'], df['Date'], label="Low")
plt.plot(df['Close'], df['Date'], label="Close")
plt.yticks(df['Date'])
plt.ylabel("Date")
plt.xlabel("Rupees")
plt.legend()
plt.show()
```

```python
# 5 Drawing three lines on same plot
marks = {
    'Name': ['Vilas', 'Abin', 'Sudheer', 'Madhan', 'Parashuram'],
    'Math': [78, 85, 90, 66, 88],
    'Science': [72, 80, 95, 70, 84],
    'English': [75, 82, 88, 69, 90]}
df = pd.DataFrame(marks)
plt.figure(figsize=(8, 4))
plt.title("Student Marks")
plt.plot(df['Name'], df['Math'], label="Maths")
plt.plot(df['Name'], df['Science'], label="Science")
plt.plot(df['Name'], df['English'], label="English")
plt.ylabel("Marks")
plt.xlabel("Students")
plt.legend()
plt.show()
```

```python
# 9 Display current axis limit and set new axis limit
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.plot(x, y)
x_limit = plt.xlim()
y_limit = plt.ylim()
print(f"X axis limit: {x_limit} \nY axis limit: {y_limit}")

# seting new axis limit
plt.xlim(0, 7)
plt.ylim(0, 40)
print("New X-axis limits:", plt.xlim())
print("New Y-axis limits:", plt.ylim())
plt.title("Set axis limit")
plt.show()

```

```python
# 10 Plot quantities having x and y positions
df = pd.DataFrame({'Height': [150, 155, 160, 165, 170, 175],
    'Weight': [50, 54, 58, 63, 68, 72]})
plt.title("Height VS Weight")
plt.plot(df['Height'], df['Weight'])
plt.xlabel('Height in cm')
plt.ylabel('Weight in KG')
plt.show()
```

```python
# 11 Plot several lines with different format styles in one command using arrays

x = np.array([1, 2, 3, 4, 5])
y1 = np.array([1, 4, 9, 16, 25])
y2 = np.array([2, 8, 18, 32, 50])
y3 = np.array([1, 3, 5, 7, 9])

# Plot multiple lines in one command with different styles
plt.plot(
    x, y1, 'r--', 
    x, y2, 'g^-',
    x, y3, 'bo:' 
)

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Multiple Lines with Different Styles')
plt.legend(labels=['vilas', 'vinay', 'kiran'])

plt.show()
```

```python
# 12 Program to create multiple types of charts
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]
categories = ['A', 'B', 'C', 'D', 'E']
values = [12, 19, 7, 15, 22]
data = [10, 12, 15, 18, 20, 22, 25, 28, 30]

# Line Chart
plt.figure()
plt.plot(x, y)
plt.title('Line Chart')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()

# Scatter Chart
plt.figure()
plt.scatter(x, y)
plt.title('Scatter Chart')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()

# Bar Chart
plt.figure()
plt.bar(categories, values)
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()

# Histogram
plt.figure()
plt.hist(data)
plt.title('Histogram')
plt.xlabel('Data values')
plt.ylabel('Frequency')
plt.show()
```

```python
# 13 Display grid and line chart of closing values
df = pd.read_csv('fdata.csv')
# Convert object to date type
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%y")
plt.title("Closing balance from 3rd to 7th October")
plt.plot(df['Close'], df['Date'])
plt.yticks(df['Date'])
plt.xlabel("Rupees")
plt.ylabel("Date")
plt.grid(linestyle="-", linewidth="0.5", c="blue")
plt.show()
```

```python
# 14 Customise grid line
df = pd.read_csv('fdata.csv')
# Convert object to date type
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%y")
plt.title("Closing balance from 3rd to 7th October")
plt.plot(df['Close'], df['Date'])
plt.yticks(df['Date'])
plt.xlabel("Rupees")
plt.ylabel("Date")
plt.grid()
# Custimize grid
plt.minorticks_on()
plt.grid(which='major', linewidth=1.2)
plt.grid(which='minor', linewidth=0.5)

# turn off ticks
plt.tick_params(axis='both', which='both', length=0)
plt.show()
```

```python
# 15 Multiple plots

df = pd.read_csv('fdata.csv')
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%y")

plt.figure(figsize=(14, 8))
# Plot 1
plt.subplot(2, 2, 1)
plt.plot(df['Open'], df['Date'])
plt.title('Plot 1: Open value')
plt.yticks(df['Date'])
plt.xlabel('Open value')
plt.ylabel('Date')

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(df['Close'], df['Date'])
plt.title('Plot 2: Close value')
plt.yticks(df['Date'])
plt.xlabel('Close value')
plt.ylabel('Date')

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(df['High'], df['Date'])
plt.title('Plot 3: High value')
plt.yticks(df['Date'])
plt.xlabel('High value')
plt.ylabel('Date')

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(df['Low'], df['Date'])
plt.title('Plot 4: Low value')
plt.yticks(df['Date'])
plt.xlabel('Low value')
plt.ylabel('Date')

# Display plots
plt.subplots_adjust(hspace=0.35, wspace=0.3)
# plt.tight_layout()
plt.show()
```
