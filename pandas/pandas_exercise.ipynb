```python
import pandas as pd
import numpy as np
```

```python
# 1D array
data = pd.Series([1, 2, 3, 4, 5, 6, 7, 8, 9])
data
```

```python
# 2 Convert pandas to List
lst = data.to_list()
lst
```

```python
# 3 Add, Subtract, Multiply and Divide two pandas series
s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 9])
print(f'Addition of s1 + s2 : \n{s1 + s2}\n')
print(f'Subtraction of s1 - s2 : \n{s1 - s2}\n')
print(f'Multiplication of s1 x s2 : \n{s1 * s2}\n')
print(f'division of s1 / s2 : \n{s1 // s2}\n')
```

```python
# 4 Powers of array values element-wise
s1 = pd.Series([0, 1, 2, 3, 4, 5, 6])
s2 = pd.Series([3, 3, 3, 3, 3, 3, 3])
print(s1 ** s2)
```

```python
# 5 Dictionary ot pandas
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael',
'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
data = pd.DataFrame(exam_data, index=labels)
data
```

```python
# 6 Summary of given data
data = pd.DataFrame(exam_data, index=labels)
data.info()
```

```python
# 7 Print first three rows of data
data = pd.DataFrame(exam_data, index=labels)
data.head(3)
```

```python
# 8 Display specified attributes
data = pd.DataFrame(exam_data, index=labels)
data[['name','score']]
```

```python
# 9 Select specified columns and rows from the given dataframe
data = pd.DataFrame(exam_data, index=labels)
data.loc[['b', 'd', 'f', 'g'], ['name', 'score']]
```

```python
# 10 Select rows where number of attempts is greater than 2
data = pd.DataFrame(exam_data, index=labels)
data[data['attempts'] > 2]
```

```python
# 11 Count Number of rows and columns
rows, cols = data.shape
print(f"rows: {rows}")
print(f"cols: {cols}")
```

```python
# 12 Select score where score is missing
data[data['score'].isna()]
```

```python
# 13 Select rows score > 15 and attempts < 2
data[(data['attempts'] < 2) & (data['score'] > 15)] 
```

```python
# 14 Change score row 'd' to 11.5
data.loc['d', 'score'] = 11.5
data
```

```python
# 15 Sum of examination attempt by the students
data['attempts'].sum() 
```

```python
# 16 Mean score for given data frame
data['score'].mean()
```

```python
# 17 Added new row to the given dataframe
data.loc['k'] = {'name' : "Suresh", 'score': 15.5, 'attempts': 1, 'qualify': "yes"}
data
```

```python
# 18 Sort dataframe by name in descending order and by score in ascending order
data.sort_values(by=['name', 'score'],ascending=[False, True])
```
