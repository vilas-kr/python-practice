```python
import numpy as np
import plotly.express as px
import pandas as pd
```

```python
# 1 Scatter plot for random 1000 x and y coordinates

x = np.random.rand(1000)
y = np.random.rand(1000)

fig = px.scatter(x=x, y=y,title="Random x and y coordinates")
fig.show()
 
```

```python
# 2 Draw line and scatter plot for 100 x and y coordinates

x = np.random.randint(1, 100, size = 100)
y = np.random.randint(1, 100, size = 100)

# Line chart
fig = px.line(x=x, y=y, title="Line plot for random x and y coordinates",
              labels={'x':'X axis', 'y':'Y axis'}, markers=True)
fig.show()

# Scatter plot
fig = px.scatter(x=x, y=y, title="Scatter plot for random x and y coordinates",
                 labels={'x':'X axis', 'y':'Y axis'})
fig.show()
```

```python
# 3 Scatter plot for random 500 x and y coordinates and style it

x = np.random.rand(500)
y = np.random.rand(500)

fig = px.scatter(x=x, y=y, title="Scatter plot with style",
                 labels={'x':'X axis', 'y':'Y axis'})
fig.update_traces(
    marker = dict(
        symbol='square',
        color="green"
    ),
)
fig.update_layout(
    xaxis = dict(
        showgrid = False,
        title_font = dict(size=14, color='blue', family="Arial"),
    ),
    yaxis = dict(
        showgrid = False,
        title_font = dict(size=14, color='blue', family="Arial"),
    ),
    title= dict(
        x=0.5,
        y=0.9,
    ),
    title_font = dict(
        family='Times New Roman',
        size = 22,
        color='red'
    ),
)
fig.show()
```

```python
# 4 Scatter plot for the given dataset

df = pd.read_csv('population.csv')

fig = px.scatter(df, x='State', y='Population',
                 color='Rank',
                 title='USA Population',
                 hover_name="State",
                 hover_data=['Postal'])
fig.show()
```
