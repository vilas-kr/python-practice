```python
# Extract data from the text file and store it in pandas
import pandas as pd
emp_data = pd.read_csv('employee_data.csv')
emp_data
```

```python
# Clean data
emp_data['Phone'] = emp_data['Phone'].astype('string')
emp_data.info()
```

```python
# Analyse data
pd.set_option('display.max_rows', None)
emp_data
```

```python
# Check null values
emp_data.isna().sum()
```

```python
# Check duplicate values
emp_data['Employee_ID'].duplicated().sum()
```

```python
# Delete duplicate record by employee id
emp_data.drop_duplicates(subset='Employee_ID', inplace=True)
emp_data['Employee_ID'].duplicated().sum()
```

```python
# Drop records where Department is null
emp_data.dropna(subset="Department", inplace=True)
```

```python
# Replace null values with average salary of that department
emp_data['Salary'] = emp_data.groupby('Department')['Salary'].transform(lambda x : x.fillna(x.mean()))
emp_data['Salary'].isna().sum()
```

```python
# Replace null values in gender with PNS (Prefer not to Say)
emp_data['Gender'] = emp_data['Gender'].fillna('PNS')
emp_data['Gender'].isna().sum()
```

```python
# Replace null values in Age column with mean age of that department
emp_data['Age'] = emp_data.groupby("Department")['Age'].transform(lambda x : x.fillna(x.mean()))
emp_data['Age'].isna().sum()
```
