```python
import pandas as pd
```

```python
lst = [1, 2, 3, 4, 5]
series = pd.Series(lst)
print(series.name)
series.name = 'numbers'
series.index = ['vilas','vinay','parashuram', 'sudheer', 'tejash']
series['parashuram':] = 1
series.drop(['vilas', 'vinay'], inplace = True)
series['kiran'] = 6
series
```

```python
# Add two series
s1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
s2 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])

s = pd.concat([s1, s2], ignore_index=True)
s
```

```python
# dictionary to series
d = {'apple': 70, 'orange': 30, 'banana': 10, 'avacado': 100, 'apple': 50}
s = pd.Series(d)
s.name = 'price'
print(s.count)
```
